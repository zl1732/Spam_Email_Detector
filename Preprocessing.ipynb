# Preprocessing

## Reading into DataFrame

```python
import numpy as np
import pandas as pd
import os
import codecs
import re
import pickle
```

```python
def cleanText(text):
    # This function takes in a text string and cleans it 
    # by keeping only alphanumeric and common punctuations
    # Returns the cleaned string
    clean_text = text.replace('\n','').replace('\r','')
    #clean_text = re.sub(r'\s*(\(\d)|(\))\s*', '', text.lower())
    clean_text = re.sub(r'[^a-zA-Z0-9.:!? ]','',clean_text)
    return clean_text
```

```python
def readEmail(path):
    # This function takes a path to an email text file
    # Returns a tuple of the subject line and the body text
    with codecs.open(path, "r",encoding='utf-8', errors='ignore') as f:
        subject = cleanText(f.readline()[9:])
        body = cleanText(f.read())
        return [subject, body]
```

```python
subjects = []
bodys = []
spam = []
# read hams
for i in range(1,7):
    for filename in os.listdir("enron%d/ham"%i):
        if filename.endswith(".txt"):
            subject, body = readEmail("enron%d/ham/"%i+filename)
            subjects.append(subject)
            bodys.append(body)
            spam.append(0)
    # read spams
    for filename in os.listdir("enron%d/spam"%i):
        if filename.endswith(".txt"):
            subject, body = readEmail("enron%d/spam/"%i+filename)
            subjects.append(subject)
            bodys.append(body)
            spam.append(1)
```

```python
data = pd.DataFrame()
data['subject'] = subjects
data['body'] = bodys
data['spam'] = spam
```

```python
df = data[['body', 'spam']]
df.columns = ['Text','Class']
with open('enron_email.df', 'wb') as handle:
    pickle.dump(df, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
